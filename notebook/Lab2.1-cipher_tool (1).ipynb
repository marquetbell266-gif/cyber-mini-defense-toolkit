The Caesar Cipher is one of the oldest encryption methods. It works by shifting each letter in a message by a fixed number of positions in the alphabet.

**Example with shift = 3:**
- `HELLO` becomes `KHOOR`
- `cyber` becomes `fbehu`

## Step 1 — Define the Caesar Encrypt Function

The function below shifts each alphabetic character by a specified value.

It handles:
- Uppercase letters
- Lowercase letters
- Spaces, numbers, and symbols by leaving them unchanged

```python
def caesar_encrypt(text, shift):
    result = ""

    for char in text:
        if char.isalpha():
            # Keep uppercase and lowercase letters in their original format
            base = 65 if char.isupper() else 97
            result += chr((ord(char) - base + shift) % 26 + base)
        else:
            # Keep spaces, numbers, and symbols unchanged
            result += char

    return result
```

## Step 2 — Encryption Test: `HELLO`

The first test encrypts the uppercase message `HELLO` using a shift value of `3`.

Expected output: `KHOOR`

```python
message_1 = "HELLO"
shift = 3

encrypted_1 = caesar_encrypt(message_1, shift)

print("Original message:", message_1)
print("Encrypted message:", encrypted_1)
```

## Step 3 — Encryption Test: `cyber`

The second test encrypts the lowercase message `cyber` using shift `3`.

Expected output: `fbehu`

```python
message_2 = "cyber"

encrypted_2 = caesar_encrypt(message_2, shift)

print("Original message:", message_2)
print("Encrypted message:", encrypted_2)
```

## Step 4 — Encryption Test: `Security 101`

The third test uses a mixed-case message with numbers and a space.

Expected output: `Vhfxulwb 101`

Notice that letters are shifted, while spaces and numbers stay the same.

```python
message_3 = "Security 101"

encrypted_3 = caesar_encrypt(message_3, shift)

print("Original message:", message_3)
print("Encrypted message:", encrypted_3)
```

## Step 5 — Define the Caesar Decrypt Function

To decrypt a Caesar Cipher message, we reverse the shift.

If encryption uses `+3`, decryption uses `-3`.

```python
def caesar_decrypt(text, shift):
    return caesar_encrypt(text, -shift)
```

## Step 6 — Decryption Test: `HELLO`

Now decrypt the encrypted version of `HELLO` and confirm that it returns to the original message.

```python
decrypted_1 = caesar_decrypt(encrypted_1, shift)

print("Encrypted message:", encrypted_1)
print("Decrypted message:", decrypted_1)
```

## Step 7 — Decryption Test: `cyber`

Now decrypt the encrypted version of `cyber`.

```python
decrypted_2 = caesar_decrypt(encrypted_2, shift)

print("Encrypted message:", encrypted_2)
print("Decrypted message:", decrypted_2)
```

## Step 8 — Decryption Test: `Security 101`

Now decrypt the encrypted version of `Security 101`.

```python
decrypted_3 = caesar_decrypt(encrypted_3, shift)

print("Encrypted message:", encrypted_3)
print("Decrypted message:", decrypted_3)
```
