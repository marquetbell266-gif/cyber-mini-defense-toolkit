## Step 1 — What Is Password Hashing?

Hashing transforms a password into a fixed-length value that does not look like the original password.

In this lab, we will use **SHA-256**, a common hashing algorithm, to demonstrate how passwords can be protected.

### Key idea
- The same password will always create the same hash.
- A small change in the password creates a very different hash.
- Hashing is safer than storing passwords directly as plain text.


## Step 2 — Import the Hashing Library

Python includes a built-in library called `hashlib` that allows us to create secure hashes.


```python
import hashlib
```

## Step 3 — Create a Password Hash

Run the cell below. When prompted, enter a password.

Try testing:
- A short password, such as `abc`
- A stronger password, such as `MyP@ss123`
- A password with numbers, such as `pass1234`


```python
password = input("Enter password: ")

hashed = hashlib.sha256(password.encode()).hexdigest()

print("Hashed password:", hashed)
```

## What Does Each Line Do?

- `import hashlib` loads Python’s built-in hashing library.
- `input(...)` waits for the user to type a password.
- `password.encode()` converts the password text into bytes so it can be hashed.
- `hashlib.sha256(...).hexdigest()` converts the password into a secure 64-character hash.
- `print(...)` displays the hashed result.

## Step 5 — Challenge: Add Simple Password Strength Logic

Now we will add basic password strength logic.

The script will label passwords shorter than 8 characters as **Weak password** and passwords with 8 or more characters as **Stronger password**.


```python
password = input("Enter password: ")

if len(password) < 8:
    print("Weak password")
else:
    print("Stronger password")

hashed = hashlib.sha256(password.encode()).hexdigest()

print("Hashed password:", hashed)
```

## Optional Extension

Try improving the password checker by adding checks for:

- Numbers
- Uppercase letters
- Special characters such as `!`, `@`, or `#`


```python
# Optional extension starter code
password = input("Enter password: ")

has_number = any(char.isdigit() for char in password)
has_uppercase = any(char.isupper() for char in password)
has_special = any(not char.isalnum() for char in password)

if len(password) >= 8 and has_number and has_uppercase and has_special:
    print("Strong password")
elif len(password) >= 8:
    print("Moderate password")
else:
    print("Weak password")

hashed = hashlib.sha256(password.encode()).hexdigest()
print("Hashed password:", hashed)
```
