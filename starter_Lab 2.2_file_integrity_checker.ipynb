## Key Concept — File Integrity

**File integrity** means making sure a file has not been changed, corrupted, or tampered with.

A hash works like a digital fingerprint for a file:

- Same file content = same hash
- Changed file content = different hash

This makes hashing useful for checking whether a file is still trustworthy.


## Step 1 — Create the Original Sample File

First, we will create a simple text file named `sample_file.txt`.

The original file will contain:

`This is my original cybersecurity file.`


```python
# Create the original sample file

with open("sample_file.txt", "w") as file:
    file.write("I love this this is awsome.")

print("sample_file.txt has been created.")
```

Run the cell below to confirm the file was created correctly.

## Step 3 — Define the Hashing Function

The function below reads the file in **binary mode** (`rb`) and generates a SHA-256 hash.

Binary mode is used because hashing algorithms work with bytes.


```python
import hashlib

def hash_file(filename):
    with open(filename, "rb") as file:
        content = file.read()
        return hashlib.sha256(content).hexdigest()
```

## Step 4 — Generate the First SHA-256 Hash

Now we will generate the hash for the original `sample_file.txt` file.


```python
file_name = "sample_file.txt"
original_hash = hash_file(file_name)

print("SHA-256 File Hash for Original File:")
print(original_hash)
```

## Step 5 — Modify the File

Now we will change one word in the file.

Original sentence:

`This is my original cybersecurity file.`

Updated sentence:

`This is my updated cybersecurity file.`

Only one word changes: `original` becomes `updated`.


```python
# Modify the file by changing one word

with open("sample_file.txt", "w") as file:
    file.write("This is my updated cybersecurity file.")

print("sample_file.txt has been modified.")
```

## Step 6 — Read the Updated File Content

Run the cell below to confirm the file was modified.


```python
# Read and display the modified file content

with open("sample_file.txt", "r") as file:
    updated_content = file.read()

print("Updated file content:")
print(updated_content)
```

## Step 7 — Generate the New SHA-256 Hash

Now generate a new hash after the file modification.


```python
updated_hash = hash_file(file_name)

print("SHA-256 File Hash After Modification:")
print(updated_hash)
```

## Step 8 — Compare the Two Hashes

If the hashes are different, that means the file changed.


```python
print("Original Hash:")
print(original_hash)

print("\nUpdated Hash:")
print(updated_hash)

if original_hash == updated_hash:
    print("\nResult: The file has NOT changed.")
else:
    print("\nResult: The file HAS changed.")
```

## What Happened?

Even though only one word changed, the SHA-256 hash changed completely.

This demonstrates the **avalanche effect**:

> A small change in the input creates a major change in the hash output.

This is why hashes are useful for checking file integrity.
