## Key Concept — What is an Anomaly?

An **anomaly** is something that looks unusual compared to the rest of the data.

Example:

`[1, 2, 2, 3, 100]`

Most values are small: `1`, `2`, `2`, and `3`.

The value `100` is very different, so it may be considered suspicious.

## Step 2 — Import Required Libraries

We will use:

- `IsolationForest` from `scikit-learn`
- `numpy` to organize our sample data


```python
from sklearn.ensemble import IsolationForest
import numpy as np
```

## Step 3 — Create Sample Login Attempt Data

The data below represents simple login attempt activity.

Most values are low, but one value is unusually high.


```python
# Sample login attempt data
data = np.array([[1], [2], [2], [3], [6], [100]])

print("Login attempt data:")
print(data)
```

## Step 4 — Build the Isolation Forest Model

The model will look for values that are different from the rest of the dataset.

`contamination=0.2` means we expect about 20% of the data to be unusual.


```python
model = IsolationForest(contamination=0.2, random_state=42)
model.fit(data)
```

## Step 5 — Generate Predictions

The model returns:

- `1` = normal
- `-1` = anomaly / suspicious


```python
predictions = model.predict(data)

print("Predictions:")
print(predictions)
```

## Step 6 — View Data with Predictions

This makes the results easier to understand.


```python
for value, prediction in zip(data.flatten(), predictions):
    label = "Suspicious / Anomaly" if prediction == -1 else "Normal"
    print(f"Value: {value} --> {label}")
```

## Step 7 — Interpret the Results

Look at the output above.

The value labeled **Suspicious / Anomaly** is the value the model thinks is unusual compared to the others.


## Cybersecurity Connection

In cybersecurity, anomaly detection can help identify:

- repeated failed logins
- unusual user behavior
- possible brute force attacks
- abnormal network traffic
- suspicious file activity

AI can help security teams review large amounts of data faster, but humans still need to review and confirm results.
